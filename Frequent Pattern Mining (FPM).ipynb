```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder # http://rasbt.github.io/mlxtend/user_guide/preprocessing/TransactionEncoder/
#association_rules Generates a DataFrame of association rules including the metrics 'score', 'confidence', and 'lift'
#apriori it is an alogorithm which means you may also like (Recommendation) and Get frequent itemsets from a one-hot DataFrame
```

```
store_data = pd.read_csv('store_data.csv',header=None)
store_data.head()
```

```
store_data[0]
```

```
items = (store_data[0].unique())#Unique items in 1st column
items
```

```
len(items)#Counting total number of unique items of 1st column
```

```
len(store_data)
```

```
item_list = []# Array (Empty)
for i in range(0,len(store_data)):#from 0 to 7501
  item_list.append([str(store_data.values[i,j])for j in range(0,20)])
item_list
#ans = store_data.values.tolist()#ans do same as item_list
# i will run row wise and j will column wise
```

```
te = TransactionEncoder()
#Encodes database transaction data alphabeticaly in form of a Python list of lists into a NumPy array.
#Using and TransactionEncoder object, we can transform this dataset into an array format suitable for typical machine learning APIs.
te_ary = te.fit(item_list).transform(item_list)
# the fit method, the TransactionEncoder learns the unique labels in the dataset, and via the transform method,
# it transforms the input dataset (a Python list of lists) into a one-hot encoded NumPy boolean array:
te_ary
```

```
te.columns_ #unique column names
```

```
df = pd.DataFrame(te_ary, columns=te.columns_)
#the unique column names that correspond to the data array can be accessed via the columns_ attribute
df
```

```
df['nan'].value_counts()#1st record don't have null value
```

```
df['nan']#null colunms exists due transaction encoder
```

```
del df['nan']
df.columns#columns length=120 which was length=121 before deletion of null column 
```

```
df
```

```
freq_items = apriori(df, min_support=0.02, use_colnames=True)
#The apriori function expects data in a one-hot encoded pandas DataFrame.
#http://rasbt.github.io/mlxtend/user_guide/frequent_patterns/apriori/
freq_items.head(10)
```

```
freq_items.count()
```

```
rules = association_rules(freq_items, metric="confidence", min_threshold=0.4)
rules.head(100)
```

```
freq_items['length'] = freq_items['itemsets'].apply(lambda x: len(x))#add length https://www.analyticsvidhya.com/blog/2020/03/what-are-lambda-functions-in-python/
#https://www.analyticsvidhya.com/blog/2021/07/most-powerful-python-functions-apply-and-lambda/
freq_items
```

```
freq_items[ (freq_items['length'] >= 2) &
                   (freq_items['support'] >= 0.005) ]
```

```
rules = association_rules(freq_items, metric="confidence", min_threshold=0.1)
#http://rasbt.github.io/mlxtend/user_guide/frequent_patterns/association_rules/
rules.head()
```

```
rules = association_rules(freq_items, metric="support", min_threshold=0.005)
rules.head()
```

```
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
rules
```

```
len(rules)
```

```
rules[ (rules['antecedent_len'] >= 1) &
       (rules['confidence'] > 0.2) &
       (rules['lift'] > 1.5) ]
```

```
rules[(rules['antecedent_len'] >= 1) &
      (rules['confidence'] > 0.2) &
       (rules['support'] > 0.01) ]
```
